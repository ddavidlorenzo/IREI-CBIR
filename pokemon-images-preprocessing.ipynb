```python
import cv2
import shutil
import os
```

```python
dir_origin = "/kaggle/input"
dir_target = "/kaggle/working/pokemon/"

width = 255

for dirname, _, filenames in os.walk(dir_origin):
    pokemon = dirname.split('/')[-1]
    print(pokemon)
    n = 0
    for filename in filenames:
        if filename.endswith(".jpg") and n <= 10:
            os.makedirs(dir_target+pokemon, exist_ok=True)
            img = cv2.imread(os.path.join(dirname, filename), cv2.IMREAD_UNCHANGED)
            if img is not None:
                y,x = img.shape[1], img.shape[0]
                height = int(y * (255/x))
                img = cv2.resize(img, (width, height))
                cv2.imwrite(os.path.join(dir_target+pokemon, filename), img)
                n += 1
```
